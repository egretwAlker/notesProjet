## 5 Bis: check how the similarity is calculated

```python
x = np.array(model['queen'])
y = np.array(model['king'])
print(np.dot(x, y)/np.sqrt(np.dot(x, x)*np.dot(y, y)))
print(model.similarity('queen', 'king'))
print(model.similarity('king', 'queen'))
```
